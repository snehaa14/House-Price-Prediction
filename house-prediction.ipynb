```python
# Building models to predict house price based on features like sqft, waterfront, etc.
# We'll use Linear Regression and Random Forest, then compare their accuracies.
```

```python

from sklearn.linear_model import LinearRegression
```

```python
# Creating a Linear Regression model instance
reg = LinearRegression()
```

```python
# Set 'price' as the label (target) and convert 'date' to binary: 1 for houses sold in 2014, else 0
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'], axis=1)
```

```python
# import another dependency to split our data into train and test.
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset: 90% for training and 10% for testing
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size=0.10, random_state=2)
```

```python
# Train the Linear Regression model on the training data
reg.fit(x_train, y_train)

# Evaluate the model on the test data and return the RÂ² score
reg.score(x_test, y_test)
```

```python
# Evaluate model performance using RÂ² score
# RÂ² = 0.732 â†’ Model explains ~73.2% of the variance in house prices
```

```python
# Applying Gradient Boosting Regressor for better performance:
# Gradient Boosting is an ensemble learning technique that builds models sequentially,
# where each new model corrects the errors of the previous ones.
# It's particularly effective for weak prediction models like linear regression,
# and works well with tabular data like house pricing.
```

```python
from sklearn import ensemble

clf = ensemble.GradientBoostingRegressor(
    n_estimators=400,
    max_depth=5,
    min_samples_split=2,
    learning_rate=0.1
)

clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)  
# Evaluate model accuracy on test data (RÂ² score)
```

```python
params = {
    'n_estimators': 400,         # Total number of boosting stages (trees)
    'max_depth': 5,              # Maximum depth of each tree
    'min_samples_split': 2,      # Minimum samples required to split a node
    'learning_rate': 0.1         # Shrinks the contribution of each tree
}

# Create an empty array to store test scores for each boosting stage
t_sc = np.zeros((params['n_estimators']), dtype=np.float64)
```

```python
# Predict house prices for the test set using the trained Linear Regression model
y_pred = reg.predict(x_test)  
```

```python
from sklearn.metrics import mean_squared_error  

# Loop through staged predictions after each boosting stage
for i, y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i] = mean_squared_error(y_test, y_pred)  
    # Store the test MSE at each stage
```

```python
testsc = np.arange((params['n_estimators'])) + 1 
# Create an array [1, 2, ..., n_estimators] for x-axis in plots
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1) 

# Plot the training error across boosting iterations
plt.plot(testsc, clf.train_score_, 'b-', label='Train set dev')  

# Plot the test error (MSE) across boosting iterations
plt.plot(testsc, t_sc, 'r-', label='Test set dev')  

plt.xlabel("Number of Estimators")
plt.ylabel("Deviance (Error)")
plt.title("Train vs Test Deviance")
plt.legend()  

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Initialize Random Forest Regressor with 400 trees and a maximum depth of 10
rf = RandomForestRegressor(n_estimators=400, max_depth=10, random_state=2)

# Train the model on the training data
rf.fit(x_train, y_train)

# Evaluate the model's performance on the test data using RÂ² score
r2 = rf.score(x_test, y_test)
print("Random Forest RÂ² Score:", r2)
```

```python
def predict_price_from_input(clf, features):
    print("Please enter the following details about the house:\n")
    
    # Readable prompts for commonly known features
    prompts = {
        'bedrooms': "Enter number of bedrooms",
        'bathrooms': "Enter number of bathrooms",
        'sqft_living': "Enter total living area (in sqft)",
        'sqft_lot': "Enter lot size (in sqft)",
        'floors': "Enter number of floors",
        'waterfront': "Does it have a waterfront view? (1 = Yes, 0 = No)",
        'view': "Enter view rating (0â€“4)",
        'condition': "Enter condition of the house (1â€“5)",
        'grade': "Enter construction grade (1â€“13)",
        'sqft_above': "Enter area above ground (in sqft)",
        'sqft_basement': "Enter basement area (in sqft)",
        'yr_built': "Enter year built (e.g. 1990)",
        'yr_renovated': "Enter year renovated (0 if never)",
        'zipcode': "Enter ZIP code (e.g. 98178)",
        'lat': "Enter latitude (e.g. 47.5112)",
        'long': "Enter longitude (e.g. -122.257)",
        'sqft_living15': "Enter avg living area of nearby 15 houses (sqft)",
        'sqft_lot15': "Enter avg lot size of nearby 15 houses (sqft)",
        'date': "Enter 1 if house sold in 2014, else 0"
    }

    values = []

    # Collect user input for each required feature
    for feature in features:
        prompt = prompts.get(feature, f"Enter value for {feature}")
        val = input(f"{prompt}: ")
        try:
            val = float(val)  # Ensure input is numeric
        except ValueError:
            print(f"âŒ Invalid input for {feature}. Please enter a numeric value.")
            return
        values.append(val)

    # Create DataFrame from user inputs for prediction
    input_df = pd.DataFrame([values], columns=features)

    
    predicted_price = clf.predict(input_df)[0]

    
    print("\nâœ… Predicted House Price: ${:,.2f}".format(predicted_price))
```

```python
features = train1.columns.tolist()
predict_price_from_input(clf, features)
```
